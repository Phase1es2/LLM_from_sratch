```python
import regex as re

RX = re.compile(r"""'(?:[sdmt]|ll|ve|re)| ?\p{L}+| ?\p{N}+| ?[^\s\p{L}\p{N}]+|\s+(?!\S)|\s+""")

s = " low low\n"
for match in RX.finditer(s):
    # print(match.group())
    print(match.group(0))

```
